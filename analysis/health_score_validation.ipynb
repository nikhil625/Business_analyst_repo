```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/customer_health_scores.csv")
```

```python
df.head()
```

```python
sns.histplot(df['health_score'], bins=50)
plt.title("Health Score Distribution")
plt.show()
```

```python
df['health_tier'].value_counts()
```

```python
df.groupby("health_tier")["health_score"].mean()
```

```python
sns.boxplot(x='health_tier', y='health_score', data=df)
plt.title("Health Score Distribution by Tier")
plt.show()
```
